```python
import pandas as pd
import numpy as np

df=pd.read_csv(r"car_prices.csv")
print(df)
```

```python
#classes values
for idx,x in enumerate(df.columns):
    print(df.columns[idx])
    print(df[x].unique())
```

```python
#dropping duplicates,NaN
df_test=df.dropna()
df_test=df_test.drop_duplicates()
df_test=df_test.reset_index()
df_test= df_test.drop('index', axis=1)
print(df_test)
```

```python
print(df_test)
```

```python
#check NaN
for idx,x in enumerate(df_test.columns):
    print(df_test.columns[idx],": ",df_test[x].isna().sum())
```

```python
#check unique
for idx,x in enumerate(df_test.columns):
    print(df_test.columns[idx])
    print(df_test[x].unique())
```

```python
grouped_df = df_test.groupby('make').agg({'sellingprice':['mean']}).reset_index()
print(grouped_df.stack())
```
