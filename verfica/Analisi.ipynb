```python
#creazione dizionari per le cat ordinabili

diz_G1_cat={'low':0,'Mid':1,'high':2}     
diz_G2_cat={'Low':0,'mid':1,'High':2} 
diz_no_yes={'no':0,'yes':1}    #schoolsup,famsup,paid,activities,nursery,higher,internet,romantic
diz_school={'GP':0,'MS':1}
diz_sex={'F','M'}
diz_address={'U':0,'R':1}
diz_famsize={'GT3':0,'LE3':1}
diz_Pstatus={'A':0,'T':1}
diz_age={'$18.0':18, '$17.0':17, '$15.0':15, '$16.0':16, '$19.0':19, '$22.0':22, '$20.0':20, '$21.0':21}  #dato che son pochi valori uso il diz al posto che lo slicinh
```

```python
#conversione a float di age e check
df['age'] = df['age'].map(diz_age)
df['age']=df['age'].astype(float)
df['age']
```

```python
#assegnazione di tutti i dizionari alle rispettive colonne

df['G1_cat'] = df['G1_cat'].map(diz_G1_cat)
df['G2_cat'] = df['G2_cat'].map(diz_G2_cat)
df['famsup'] = df['famsup'].map(diz_no_yes)
df['schoolsup'] = df['schoolsup'].map(diz_no_yes)
df['romantic'] = df['romantic'].map(diz_no_yes)
#df['sex'] = df['sex'].map(diz_sex)           da errore per qualche motivo
df['school'] = df['school'].map(diz_school)
df['paid'] = df['paid'].map(diz_no_yes)
df['activities'] = df['activities'].map(diz_no_yes)
df['nursery'] = df['nursery'].map(diz_no_yes)
df['higher'] = df['higher'].map(diz_no_yes)
df['internet'] = df['internet'].map(diz_no_yes)
df['address'] = df['address'].map(diz_address)
df['famsize'] = df['famsize'].map(diz_famsize)
df['Pstatus'] = df['Pstatus'].map(diz_Pstatus)

```

```python
#prendiamo solo le colonne con correlazione maggiore >= a 0.25 o <=-0,25 che sono quelle più efficaci per il nostro modello

df_fine=df[['G3','G1_cat','G2_cat','higher','failures','school']].copy()
df_fine
```

#

```python
#droppiamo in na per il modello

df_fine['G1_cat']=df_fine['G1_cat'].fillna(df_fine['G1_cat'].mode()[0])
df_fine['G2_cat']=df_fine['G2_cat'].fillna(df_fine['G2_cat'].mode()[0])
df_fine['higher']=df_fine['higher'].fillna(df_fine['higher'].mode()[0])
df_fine['failures']=df_fine['failures'].fillna(df_fine['failures'].mode()[0])
df_fine['school']=df_fine['school'].fillna(df_fine['school'].mode()[0])
df_fine.isna().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()
X = df_fine[['G1_cat','G2_cat','higher','failures','school']]
y = df_fine['G3']

model.fit(X, y)

predictions = model.predict(X)
r2 = r2_score(y,predictions)
print(r2)

```

```python
df_comparazione=df_fine[['G3']]
df_comparazione['predizioni']=predictions
df_comparazione
```

```python
plt.figure(figsize=(6,6))
plt.scatter(y, predictions, c='crimson')

p1 = max(max(predictions), max(y))
p2 = min(min(predictions), min(predictions))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```
